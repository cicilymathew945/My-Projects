### K-MODES

```python
import pandas as pd
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt

# 1. Prepare your data (using the filtered segmentation dataframe)
# Ensure all columns are treated as strings/categories for K-Modes
df_kmodes = scaled_data.astype(str) 

# 2. Find optimal number of clusters (Elbow Method)
cost = []
K = range(1, 8)
for k in K:
    km = KModes(n_clusters=k, init='Huang', n_init=10, verbose=0)
    km.fit(df_kmodes)
    cost.append(km.cost_)

# Plot the Elbow
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of Clusters (k)')
plt.ylabel('Cost (Dissimilarity)')
plt.title('Elbow Method for Optimal K (K-Modes)')
plt.show()
```

```python

# 3. Run the final K-Modes model
# Let's assume the elbow is at 4 clusters
optimal_k = 5
km = KModes(n_clusters=optimal_k, init='Huang', n_init=10, verbose=1)
clusters = km.fit_predict(df_kmodes)

# 4. Add cluster labels back to your original data
df_segmentation['Segment_Labels'] = clusters
```

### HIERARCHICAL CLUSTERING


```python
# CLUSTERING CODE
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
import matplotlib.pyplot as plt

# 1. Use your already MinMax scaled dataframe
# df_scaled = (your scaled dataframe here)

# 2. Perform Linkage using Ward's Method
# Ward's minimizes the variance within clusters
Z = linkage(scaled_data, method='ward')

# 3. Plot the Dendrogram
plt.figure(figsize=(15, 7))
plt.title('Dendrogram for Boater Segmentation (MinMax Scaled)')
plt.xlabel('Respondents')
plt.ylabel('Distance (Ward)')

# Truncate to make it readable if you have many rows
dendrogram(Z, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=10.)
plt.axhline(y=15, color='r', linestyle='--') # Example horizontal line to cut the tree
plt.show()

```

```python
# 4. Extract Cluster Labels
# Based on the dendrogram, choose k (number of segments)
k = 5
df_segmentation['Segment_Labels'] = fcluster(Z, k, criterion='maxclust')
```

```python
# RESULTS AND VISUALIZATION
from sklearn.metrics import euclidean_distances
import numpy as np

def calculate_density_metrics(data, labels):
    unique_labels = np.unique(labels)
    metrics = []

    for label in unique_labels:
        # Filter data for this specific cluster
        cluster_points = data[labels == label]
        
        # Calculate all pairwise distances within the cluster
        dists = euclidean_distances(cluster_points)
        
        # Diameter: Max distance between any two points
        diameter = np.max(dists)
        
        # Density: Average distance between all points
        avg_dist = np.mean(dists[np.nonzero(dists)]) # exclude self-distance (0)
        
        # Sample Count
        count = cluster_points.shape[0]
        
        metrics.append({
            'Cluster': label,
            'Count': count,
            'Avg_Intra_Dist': round(avg_dist, 3),
            'Diameter': round(diameter, 3)
        })
        
    return pd.DataFrame(metrics)

# Run the validation
# Use your MinMax scaled data and the labels from the Dendrogram
density_report = calculate_density_metrics(scaled_data, df['Segment_Labels'])
print(density_report)
```

## K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Prepare your data 
# Ensure data is numerical and scaled (e.g., your 'scaled_data' from before)
df_kmeans = scaled_data.copy() 

# 2. Find optimal number of clusters (Elbow Method)
inertia = []
K = range(1, 10)

for k in K:
    # random_state ensures you get the same results every time you run it
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
    kmeans.fit(df_kmeans)
    inertia.append(kmeans.inertia_)

# 3. Plot the Elbow
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'ro-') # 'ro-' creates a red line with dots
plt.xlabel('No. of Clusters (k)')
plt.ylabel('Inertia (Within-Cluster Sum of Squares)')
plt.title('Elbow Method for Optimal K (K-Means)')
plt.grid(True)
plt.show()
```

```python
# 3. Run the final K-means model
# Let's assume the elbow is at 4 clusters
optimal_k = 3
km = KMeans(n_clusters=optimal_k, init='k-means++', n_init=10,random_state=42, verbose=1)
clusters = km.fit_predict(df_kmeans)

# 4. Add cluster labels back to your original data
df_segmentation['Segment_Labels'] = clusters
```

```python
from sklearn.metrics import silhouette_score, silhouette_samples
import numpy as np

# Assuming 'labels' are your K-Means results and 'df_weighted' is your data
score = silhouette_score(df_segmentation, clusters)
print(f"Overall Silhouette Score: {score:.4f}")

# TIP: If your score is near 0.12, it matches your 12% varianceâ€”meaning 
# clusters are heavily overlapping.
```
